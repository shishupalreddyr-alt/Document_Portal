### Retreival process
#Let us fetch the relevant content or rank the retreived results from the vector store using similarity search

```python
relevant_docs=faiss_db.similarity_search("What is the availability of Joule? in SP Products portfolio", k=5) 
relevant_docs
```

## As a Retreiver 
#Conversion of Vector stores into retreiver class, this can be used later in retriver process

```python
retriever=faiss_db.as_retriever(search_kwargs={"k":5})
retriever.invoke("What is the availability of Joule? in SP Products portfolio")

```

#### Similarity Search with score
There are some FAISS specific methods. One of them is similarity_search_with_score, which allows you to return not only the documents but also the distance score of the query to them. The returned distance score is L2 distance. Therefore, a lower score is better.

```python
### querying 
query="What is SAP Joule availability?"

docs_with_score=faiss_db.similarity_search_with_score(query)
docs_with_score
```

### Question: user question
Context: based on the question retrieving the info from the vector database

```python
prompt_template="""
You are a helpful assistant. Based on the following context, answer the question.
"Important"
 Should the response is not clear or not known, please ask for clarification or respond with
 "Please share more information/context for your question."
Context: {context}

Question: {question}

Answer:
"""
```

```python
from langchain_core.prompts import PromptTemplate
```

```python
prompt=PromptTemplate(
    template=prompt_template,
    input_variables=["context", "question"]
)
```

```python
prompt
```

```python
from langchain_core.output_parsers import StrOutputParser

outputparser=StrOutputParser()
```

```python
def generate_response(context, question):
    prompt = PromptTemplate(
        template=prompt_template,
        input_variables=["context", "question"]
    )
    output_parser = StrOutputParser()
    # Further implementation...
    prompt_text = prompt.format(context=context, question=question)
    response = output_parser.parse(prompt_text)
    return response
```

```python
def format_docs(docs):
    return "\n\n".join([doc.page_content for doc in docs])
    
```

```python
from langchain_core.runnables import RunnablePassthrough
```

```python
##Langchain Expression Language  LCEL
rag_chain = (
    {"context": retriever | format_docs, "question": RunnablePassthrough()}
    | prompt
    | llm
    | outputparser
)
```

```python
rag_chain.invoke("Tell me about the key features of SAP Joule")
```
